# Importing Libraries and dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import os
```

```python
path = r'C:\Users\ranji\Boat Analysis'
df = pd.read_csv(os.path.join(path, 'Data', 'Prepared Data', 'boats_clean.csv'), index_col = 0)
```

```python
df.head()
```

# Correlation 1 - Heat Map

```python
df.columns
```

```python
# Creating subset with numerical columns only
df_cor = df[['Year Built','Price (EUR)', 'Length', 'Width', 'Number of views last 7 days']]
```

```python
f,ax = plt.subplots(figsize=(10,10))
corr = sns.heatmap(df_cor.corr(), annot = True, ax = ax)
```

```python
heat_map = df_cor.value_counts().plot.bar()
```

Correlation Analysis:
We see a strong positive correlation relationship between Price with Length and Price with Width.

# Correlation 2 - Scatter Plot

```python
sns.lmplot(x = 'Length', y = 'Price (EUR)', data = df)
```

```python
sns.lmplot(x = 'Width', y = 'Price (EUR)', data = df)
```

Correlation Analysis: We see a very similar trend in both plots with the Length and Width. The price of the boats do increase with the increase in Length and Width however there is a "sweet spot" where we see the highest prices of boats have a length between 25m and 40m and a Width between 6m and 8m. 

There are a few outliers with low prices and high length and widths however this can be because of the boats being very old with old technology. 

# Correlation 3 - Pair Plot
